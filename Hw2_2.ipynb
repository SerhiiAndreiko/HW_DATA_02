# **Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць. Використовуйте її для методу apply та створіть новий стовпчик "avg"**

```python

# Визначення функції fill_avg_salary
fill_avg_salary = {'Зарплата на місяць': 'median'}
result = df.groupby(['Посада']).agg(fill_avg_salary)
print(result)

def fill_avg_salary(row):
    return np.mean(row['Зарплата на місяць'])

# Застосування функції за допомогою методу apply
df['avg'] = df.apply(fill_avg_salary, axis=1)
result = df.groupby('Посада')['Зарплата на місяць'].agg(avg= 'mean').reset_index()

# Виведення результатів
print(result)
```
